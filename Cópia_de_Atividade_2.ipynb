```python
import pandas as pd

# Carregar o arquivo CSV
file_path = '/content/estacoesbike.csv'
df = pd.read_csv(file_path, delimiter=';')

# Calcular Total de Estações
total_estacoes = df['nome'].count()

# Calcular Capacidade Total
capacidade_total = df['capacidade'].sum()

# Calcular Quantidade de Bairros distintos
quantidade_bairros = df['bairro'].nunique()

# Criar um dataframe com os resultados
resultados = pd.DataFrame({
    'Métrica': ['Total de Estações', 'Capacidade Total', 'Quantidade de Bairros distintos'],
    'Valor': [total_estacoes, capacidade_total, quantidade_bairros]
})

# Exibir o dataframe
print(resultados)
```
