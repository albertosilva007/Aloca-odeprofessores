```python
import pandas as pd
import numpy as np
```

Analisar as infrações de trânsito em Recife durante o ano de 2024 para identificar padrões e áreas de maior incidência.

```python
df = pd.read_csv(" http://dados.recife.pe.gov.br/dataset/6399f689-f1a7-453b-b839-413bd665c355/resource/4adf9430-35a5-4e88-8ecf-b45748b81c7d/download/infracoestransparencia-janeiro-a-outubr", delimiter=';')
```

Exibir as primeiras linhas do DataFrame para entender a estrutura dos dados

```python
print(df.head())
```

Análise e Visualização dos Dados

Análise das infrações por bairro

```python

print(df.columns)




```

```python
infracoes_por_bairro = df['localcometimento'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 8))
infracoes_por_bairro.plot(kind='barh', color='lightgreen')
plt.title('Top 10 Bairros com Mais Infrações de Trânsito (2024)')
plt.xlabel('Número de Infrações')
plt.ylabel('Bairro')
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
